## Price-Trend Time-Series Analysis

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARMA
import statsmodels.api as sm

import warnings
warnings.filterwarnings('ignore', category=RuntimeWarning)
```

```python
# lets get the data first
df = pd.read_csv('data/Price.csv')
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df.date)
```

```python
df1 = df.set_index('date')
df1.head()
```

```python
# How is our data looking?
plt.plot(df1)
```

```python
# let make a function to test the statinarity of the data
def test_stationarity(timeseries):
    Rolmean = timeseries.rolling(window=4).mean()
#     Rolstd = timeseries.rolling(window=4).std()
    
    orig = plt.plot(timeseries, color = 'blue', label = "Orginal")
    mean = plt.plot(Rolmean, color = 'red', label = "Rolling mean")
#     std = plt.plot(Rolstd, color = 'green', label = "Rolling std")
    plt.legend(loc='best')
    plt.show()
    
    print('Result of Dickey-Fuller Test:')
    dftest = adfuller(timeseries['price'], autolag ='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', 'Lags used', "Observations"])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' %key] = value
    print(dfoutput)
```

```python
# Is the mean value stationary?
test_stationarity(df1)
```

```python
# Can we transform the data to obtain the stationarity? #lets start with log transformation
log_df = np.log(df1)
plt.plot(log_df)
```

```python
# Well slightly better but not perfect. Lets check the stationarity stats
test_stationarity(log_df)
```

### Hmmmm... has not improved much? what's next?

```python
# Let try to nullify the fluctuation by substracting it by moving avg

movingAvg = log_df.rolling(window=4).mean()
logAvg_df = log_df -movingAvg
plt.plot(logAvg_df)
```

```python
logAvg_df.dropna(inplace=True)
```

```python
test_stationarity(logAvg_df)
```

```python
logAvg_df.index = pd.DatetimeIndex(logAvg_df.index).to_period('M')
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)

fig = sm.graphics.tsa.plot_acf(logAvg_df['price'], lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(logAvg_df['price'], lags=40, ax=ax2)
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(logAvg_df, order=(2, 0, 2))
result = model.fit()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
result.plot_predict(start=0, end= 70, ax=ax)
plt.grid(linestyle="--")
plt.xlabel('Time', fontsize=15)
plt.ylabel('Airbnb Price (log)', fontsize=15)
plt.savefig('Fig/PriceTrend.png')
plt.show()
```

```python
sum((result.fittedvalues-logAvg_df['price'])**2)
```

```python
result.summary()
```

```python
residuals = pd.DataFrame(result.resid)
residuals.plot()
```

```python
residuals.plot(kind="kde")
plt.show()
print(residuals.describe())
```
